Currently, I am doing some of the sorting and condensing of my files in my streamlit app, which is not good since I can't actually push my larger files to the github repo and because I don't need all the information in those files, just the information for my visualizations

1st to fix: final-project-data2.csv which is use for my first two figures on the basic pageview data page of my streamlit app

```python
import pandas as pd
```

```python
df = pd.read_csv("final-project-data2.csv")
```

```python
countries = ["US", "JP", "IN", "DE", "GB"]

countrydf = df.sort_values(by='date', ascending=False)
countrydf = countrydf.groupby(["date", "country_code",])['pageviews'].sum().reset_index()

countrydf["date"] = pd.to_datetime(countrydf["date"])
```

```python
countrydf.to_csv("basic_pageview_data_fig1,2.csv")
```
