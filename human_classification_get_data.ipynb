```python
def combine_text(row):
    # Make sure attributes is a dictionary
    attributes = row["attributes"]
    if not isinstance(attributes, dict):
        attributes = {}
    
    # Get occupation safely
    occupation = attributes.get("occupation", "")
    
    # Combine text
    return f"{row['label']} {row['description']} {occupation}"
```

```python
df_humans["summary"] = df_humans.apply(combine_text, axis=1)
```

```python
df_humans.head()
```

```python
import torch
import pandas as pd
from transformers import pipeline
from tqdm.notebook import tqdm
```

```python
MODEL_NAME = "facebook/bart-large-mnli"
DEVICE = 0 if torch.cuda.is_available() else -1

print(f"Loading model: {MODEL_NAME} on device: {'GPU' if DEVICE == 0 else 'CPU'}")
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model=MODEL_NAME,
    device=DEVICE,
    batch_size=32
)

classifier
```

```python
labels = ["pop culture", "historical"]
```

```python
df_humans.head()
```

```python
test = df_humans.head(10)['summary'].tolist()
```

```python
results = classifier(
        df_humans['summary'].tolist(),
        candidate_labels=labels,
        hypothesis_template= "This text is about {}.",
        multi_label=False
    )

print("Classification complete.")
```

```python
results
```

```python
len(results)
```

```python
df_humans["category"] = [r['labels'][0] for r in results]
```

Now I have labelled all of my human qids as pop culture or not. Next I want to extract the labels and use them with other dataframes to create some visualizations

```python
df_humans = df_humans.rename(columns={'QID': 'qid'})
```

```python
human_data = pd.read_csv("human-data.csv")
```

```python
human_data.head()
```

```python
merged = human_data.merge(df_humans[["qid", "category"]], on="qid", how="left")
```

```python
merged.tail()
```

Now I have my dataframe with the columns I need. I can use this to create my visualizations in my app

```python
merged.to_csv("human_classification.csv", index=False)
```

```python
merged['date'] = pd.to_datetime(merged['date'])
```

```python
# Group by date and predicted_category
merged_pct = merged.groupby(['date', 'category', 'country_code'])['pageviews'].sum().reset_index()

merged_pct['date'] = pd.to_datetime(merged_pct['date'])


# Pivot so that each category is a column
merged_daily_category_pivot = merged_pct.pivot(index=['date', 'country_code'], columns='category', values='pageviews').fillna(0)

# Calculate percentages
merged_daily_category_pct = merged_daily_category_pivot.div(merged_daily_category_pivot.sum(axis=1), axis=0)
```
